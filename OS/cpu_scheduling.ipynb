```python


def calculate_fcfs(process):
    #frist come first serve:
    process.sort(key=lambda x: x['arrival'])
    
    current_time = 0
    total_time = 0
    total_wating_time = 0
    
    for p in process:
        if current_time<p['arrival']:
            current_time = p['arrival']
        
        p['completion_time'] = current_time + p['burst']
        p['turnaround_time'] = p['completion_time'] - p['arrival']
        p['wating_time'] = p['turnaround_time'] - p['burst']
        
        total_time += p['turnaround_time']
        total_wating_time += p['wating_time']
        current_time = p['completion_time'] 
    
    #sort back for ordering printing:
    process.sort(key=lambda x:x['id'])
    
    print("\n" + "-" * 70)
    print(f"{'PID':<12}{'Arrival':<12}{'Burst':<12}{'Completion':<12}{'TurnAround':<12}{'Wating':<12}")
    print("\n" + "-" * 70)
    
    # pring the value:
    for p in process:
        print(f"{f"{p["id"]}":<12}{f"{p["arrival"]}":<12}{f"{p['burst']}":<12}{f"{p["completion_time"]}":<12}{f"{p["turnaround_time"]}":<12}{ f"{p["wating_time"]}":<12}")
    
    n = len(process)
    print(f"Average Turnaround Time: {total_time / n:.2f}")
    print(f"Average Waiting Time   : {total_wating_time / n:.2f}")


process_data = [
        {'id': 1, 'arrival': 0, 'burst': 4},
        {'id': 2, 'arrival': 1, 'burst': 3},
        {'id': 3, 'arrival': 2, 'burst': 1},
        {'id': 4, 'arrival': 3, 'burst': 2},
        {'id': 5, 'arrival': 6, 'burst': 5}
    ]
calculate_fcfs(process_data)

```
