```python

import  numpy  as np 
import pandas as pd 
```

```python

houses = pd.read_csv(filepath_or_buffer="dataset/kc_house_data.csv")
houses.head(2)
```

```python

# minimum in a specific column:
houses["price"].min()
```

```python

# minimum value for the whole dataFrame:
houses.min()
```

```python

# similary maximum of the dataFrame:
houses.max()
```

```python

# remember we are getting a serics not a dataFrame:
type(houses.min())
```

```python

# sum in pandas: 
# Default for both numeric and object: 
# here, string are concatenating like: "yasin" + "arafat"  = yasinarafat
houses.sum()
```

```python

# If we want only integer value only then,
houses.sum(numeric_only=True) 

```

```python

# see the sum on first 10 rows:
houses.head(10).sum(numeric_only=True)
```

```python

#count function for couting values:
df = pd.read_csv(filepath_or_buffer="dataset/netflix_titles.csv",sep="|")
df.count()
```

```python

# mean: 
houses["bathrooms"].mean()
```

```python

# counting mean value only for numeric column:
houses.mean(numeric_only=True)
```

```python

# median # 3 er besi ase and 3 er kom ase
print(houses["bedrooms"].min())
print(houses["bedrooms"].median())
print(houses["bedrooms"].max())
```

```python

# mode: (average value)
houses.mode()
```

```python

# we have no object type data: 
print(houses.dtypes)
# describe:
houses.describe()
```

```python

# in this case, we will use the describe()
df.dtypes
```

```python

#default: 
df.describe(include=["int64"])
```

```python

# we can apply only for object dtype:
#unique: each string different with other
#top: most frequent string 
#freq: how many time the top string occur

df.describe(include=["object"])
```

```python

# for all types:
# both for object and int: 
df.describe(include=["object","int64"])
```
