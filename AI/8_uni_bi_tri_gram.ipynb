```python


import numpy as np 
import pandas as pd 

from sklearn.feature_extraction.text import CountVectorizer
```

```python

dataset =  {'text':['People Love To Watch Television',
    'Television is Not Good For People',
    'Lazy People Thinks Television Is Good'], 
    'output':[1,0,1]}
    

df = pd.DataFrame(data=dataset)
df
```

```python

cv = CountVectorizer()
bow = cv.fit_transform(df["text"])

cv.vocabulary_
```

```python

# people: 
bow[0].toarray()
```

```python


# bigram:
cv3 = CountVectorizer(ngram_range=(2,2)) 
bow = cv3.fit_transform(df['text'])
print(df['text'][0])
print(cv3.vocabulary_)
bow[0].toarray()

```

```python


# unigram+bigram
cv3 = CountVectorizer(ngram_range=(1,2))
bow = cv3.fit_transform(df['text'])
print(df['text'][0])
print(cv3.vocabulary_)
bow[0].toarray()

```
