# `# Bisection Method:`

<br>
<br>

```python

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python

dataframe = pd.DataFrame(columns=["a", "b", "f(a)", "f(b)", "root= (a+b)/2", "f(c)"])

def fn(x):
    return 3 * x - np.cos(x) - 1

def BisectionMethod(a, b,i, tol=1e-6):
    if fn(a) * fn(b) >= 0:
        return "Try another value of a and b"
    
    i = i + 1 
    c = (a + b) / 2
    function_a = fn(a)
    function_b = fn(b)
    function_c = fn(c)
        
    # Insert values into DataFrame
    val = {"a": a, "b": b, "f(a)": function_a, "f(b)": function_b, 
               "root= (a+b)/2": c, "f(c)": function_c}
    dataframe.loc[i] = val
        
    
    if abs(b - a) < tol or abs(function_c) < tol:
            return c
        
    if function_a * function_c > 0:
            return BisectionMethod(c,b,i)
            
    else:
        return BisectionMethod(a,c,i)
    


# Set the number of iterations
iteration = 0
root = BisectionMethod(0, 1, iteration)
print(f"After {iteration} iterations the root is: {root}")
```
